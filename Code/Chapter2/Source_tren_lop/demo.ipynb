# Cộng, trừ Matrix

```python
import numpy as np

A = np.array([[1, -4], [-4, 1]])
B = np.array([[2, -8], [-8, 2]])

print("A + B =", A + B)
print("A - (1/2)B =", A - B/2)
print("5A =", 5*A)
```

# Nhân 2 matrix

```python
import numpy as np

A = np.array([[2, 3, 1], [4, 5, 1]])
B = np.array([[-1, 2], [4, -2], [3, 6]])

print("AB =", np.dot(A, B))
print("BA =", np.dot(B, A))
```

# Ma trận nghịch đảo


```python
import numpy as np

A = np.array([[1, 3], [5, 7]])
B = np.array([[4, -6], [-8, 12]])

detA = np.linalg.det(A)  #Định thức
Ainv = np.linalg.inv(A)
detB = np.linalg.det(B)

print("Determinant of A =", detA)
print("Inverse of A =", Ainv)
print("Determinant of B =", detB)
```
