## Xử lý dữ liệu trùng lặp

```python
import pandas as pd

df = pd.DataFrame({'c1':['a', 'a', 'b', 'a', 'b'], 
                    'c2':[1, 1, 1, 2, 2],
                    'c3':[1, 1, 2, 2, 2]})
print(df)
```

```python
df_dup = df.duplicated()
print(df_dup)
```

```python
col_dup = df['c2'].duplicated()
print(col_dup)
```

```python
df2 = df.drop_duplicates()
print(df2)
```

```python
df3 = df.drop_duplicates(['c2', 'c3'])
print(df3)
```
